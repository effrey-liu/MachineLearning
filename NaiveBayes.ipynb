```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB,GaussianNB
```

```python
string1 = "Life is short, I need my love"
string2 = "I love doubleone, she makes me happy"
vect = CountVectorizer()

matrix = vect.fit_transform([string1,string2])
# type(matrix) 
# matrix是m*n的矩阵，m = 传入的string个数，n = words的个数（words.size()）
words = vect.get_feature_names()
# type(words)
# print(matrix,words)
```

```python
newsgroups = fetch_20newsgroups("./data")
# newsgroups.target_names
# newsgroups.data
x_train,x_test,y_train,y_test = train_test_split(newsgroups.data,newsgroups.target,test_size=0.25)
vect_news = CountVectorizer()
x_train = vect_news.fit_transform(x_train)
x_test = vect_news.transform(x_test)

mulNB_news = MultinomialNB()
mulNB_news.fit(x_train,y_train)
mulNB_news.score(x_test,y_test)
```
